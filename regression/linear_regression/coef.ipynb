determine coefficients (although this example is useless...)

```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
from sklearn import linear_model
%matplotlib inline
```

```python
# create data
def f(x):
    return 2. * np.sin(x) + 1.

def make_data(x, f, ep):
    return f(x) + np.random.normal(0, ep, x.shape[0])
```

```python
x = np.arange(-3, 3, 0.1)
tx = x
y = make_data(x, f, 0.1)
```

```python
plt.scatter(x, y)
```

```python
# convert x to sin(x)
x = np.sin(x)
```

```python
plt.scatter(x, y)
```

```python
x = x[:, np.newaxis]
y = y[:, np.newaxis]
```

```python
# define model
model = linear_model.LinearRegression()
model.fit(x, y)
```

```python
# view model
px = np.arange(-1, 1, 0.01)[:, np.newaxis]
py = model.predict(px)
plt.scatter(x, y, color='red')
plt.plot(px, py, color='blue')
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
ppx = np.arange(-3, 3, 0.01)[:, np.newaxis]
ppxx = np.sin(ppx)
ppy = model.predict(ppxx)
```

```python
plt.scatter(tx, y, color='red')
plt.plot(ppx, ppy, color='blue')
```
