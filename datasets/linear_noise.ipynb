# 1d data with noise
1d data with Gaussian noise  

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn import linear_model
import pandas as pd
%matplotlib inline
```

```python
a = 0.5
b = 3
def linear_func(x):
    return a * x + b
```

```python
def linear_noise(x, f, ep):
    '''add gaussian noise N(0, ep) to data x
    x is array
    ep is float value
    '''
    return f(x) + np.random.normal(0, ep, x.shape[0])
```

```python
# sample
x = np.arange(0, 10, 0.1)
y = linear_noise(x, linear_func, 0.1)
```

```python
plt.plot(x, y)
```

```python
# linear regression
model = linear_model.LinearRegression()
model.fit(pd.DataFrame(x), pd.DataFrame(y))
```

```python
y_test = model.predict(x[:, np.newaxis])
```

```python
plt.plot(x, y)
plt.plot(x, y_test, color='red')
```

```python
print(model.coef_)
print(model.intercept_)
```
