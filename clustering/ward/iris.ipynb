# hierarchical clustering
clustering by ward method  
reference: https://gomiba.co.in/blog/archives/705

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
%matplotlib inline
```

```python
# datasets
iris = load_iris()
data = pd.DataFrame(iris["data"])
data.columns = iris.feature_names
iris_labels = list(map(lambda t: iris["target_names"][t], iris["target"]))
```

```python
# clustering
cluster = linkage(data, method='ward', metric='euclidean')
```

```python
# visualize
plt.figure(num=None, figsize=(16, 9), dpi=200, facecolor='w', edgecolor='k')
dendrogram(cluster, labels=iris_labels)
plt.show()
```

```python
threshold = 0.7 * np.max(cluster[:, 2])
clustered = fcluster(cluster, threshold, criterion='distance')
```
