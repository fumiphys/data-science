# Function Fitting by Tensorflow.keras
fit sin(x) by tf.keras

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# make data
def f(x):
    return np.sin(x)

def make_data(x, f, ep):
    return f(x) + np.random.normal(0., ep, x.shape[0])

t_x = np.arange(-2. * np.pi / 3., 2. * np.pi / 3., 0.05)
t_y = make_data(t_x, f, 0.05)
plt.plot(t_x, t_y)
len(t_x)
t_x = t_x[:, np.newaxis]
t_y = t_y[:, np.newaxis]
```

```python
# define model
model = keras.models.Sequential([
    keras.layers.Dense(5, activation=tf.nn.sigmoid, input_shape=(1,)),
    keras.layers.Dense(1, activation=None)
])

model.compile(loss='mse', optimizer='adam', metrics=['mae'])
model.summary()
```

```python
# train
history = model.fit(t_x, t_y, epochs=2000, validation_split=0.2)
```

```python
history.history
plt.plot(np.arange(0, len(history.history['loss'])), history.history['loss'], color='blue')
plt.plot(np.arange(0, len(history.history['loss'])), history.history['val_loss'], color='red')
```

```python
# prediction
prediction = model.predict(t_x)
```

```python
plt.plot(t_x, t_y)
plt.plot(t_x, prediction, color='red')
```

```python
# add layer
model = keras.models.Sequential([
    keras.layers.Dense(5, activation=tf.nn.sigmoid, input_shape=(1,)),
    keras.layers.Dense(5, activation=tf.nn.sigmoid),
    keras.layers.Dense(1, activation=None)
])

model.compile(loss='mse', optimizer='adam', metrics=['mae'])
```

```python
history = model.fit(t_x, t_y, epochs=1000, validation_split=0.2, verbose=0)
```

```python
plt.plot(np.arange(0, len(history.history['loss'])), history.history['loss'], color='blue')
plt.plot(np.arange(0, len(history.history['loss'])), history.history['val_loss'], color='red')
```

```python
# prediction
prediction = model.predict(t_x)
```

```python
plt.scatter(t_x, t_y)
plt.plot(t_x, prediction, color='red')
```
